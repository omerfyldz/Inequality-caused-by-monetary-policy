## Paper Aim

CGKS ask whether monetary policy shocks cause changes in U.S. inequality. The identification problem is that the Federal Reserve reacts to the economy, so simple correlations between rates and inequality are not causal. The paper therefore uses Romer-Romer monetary policy shocks, then estimates dynamic responses with local projections.
